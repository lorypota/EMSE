### Research Question popular words

```python
from collections import Counter
import re

# Get all messages, drop any null values
messages = pr_commits_df['message'].dropna()
details_messages = pr_commit_details_df['message'].dropna()

# Combine all messages into words
all_words_commits = []
for message in messages:
    # Convert to lowercase and split into words
    # Remove punctuation and keep only alphanumeric words
    words = re.findall(r'\b[a-z0-9]+\b', message.lower())
    all_words_commits.extend(words)

all_words_details = []
for message in details_messages:
    # Convert to lowercase and split into words
    # Remove punctuation and keep only alphanumeric words
    words = re.findall(r'\b[a-z0-9]+\b', message.lower())
    all_words_details.extend(words)

# Count word frequencies
word_counts_commits = Counter(all_words_commits)
word_counts_details = Counter(all_words_details)
word_counts_combined = word_counts_commits + word_counts_details

# Get sorted list (most to least common)
most_common_words_commits = word_counts_commits.most_common()
most_common_words_details = word_counts_details.most_common()
most_common_words_combined = word_counts_combined.most_common()

print(most_common_words_combined[-20:20])
```

```python
# Common English stop words
stop_words = {'the', 'a', 'an', 'and', 'or', 'but', 'in', 'on', 'at', 'to', 'for',
              'of', 'with', 'is', 'was', 'are', 'be', 'been', 'from', 'by', 'as'}

not_interesting_words = {} #{'co', 'authored', 'github'}

filtered_words = [(word, count) for word, count in most_common_words_combined
                  if word not in stop_words
                  and word not in not_interesting_words
                  and not word.isdigit()]

#filtered_words
```

```python
least_common_words = filtered_words[::-1]
print(least_common_words[:20])
```
